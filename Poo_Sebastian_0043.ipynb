# lenguaje python, crear la clase "animal" con los atributos "nombre", "raza" crear el contructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor (__init__) inicializa los atributos del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! 🐱‍👤")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} (raza {self.raza}) está corriendo velozmente.🐱‍🏍")

    def morder(self):
        print(f"{self.nombre} (raza {self.raza}) está mordiendo a la vecina 🤦‍♀️")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Tyson", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Peques", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("Solovino", "Chafa")

# --- Uso de las funciones ---

print("--- Interacción con el Pitbull ---")
pitbull.ladrar()
pitbull.correr()

print("\n--- Interacción con el Chihuahua ---")
chihuahua.ladrar()
chihuahua.correr()

print("\n--- Interacción con el perro callejero ---")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()

print(" Programa creado por Sebastian Farfan NL = 16")
```
